```python
#importer les bibloithèques nécessaires
#Dans jumia tout le code est controlé par JavaScript
from selenium import webdriver
from time import sleep
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
```

```python
# Créer une instance de ChromeDriver
driver = webdriver.Chrome()
#Accédez à l'url ci-dessous
url = 'https://www.jumia.ma/catalog/?q=pc-portable&price=2956-36500#catalog-listing'
driver.get(url)
sleep(3)
```

```python
#find the articles
articles = driver.find_elements(By.XPATH,'//article[contains(@class, "prd _fb col c-prd") or contains(@class,"prd _fb _spn c-prd col")]')

#how many article do we have
len(articles)
```

```python
#Catching the last element
for article in articles :
    pass
```

```python
#Getting the link of the article
article_link = article.find_element(By.TAG_NAME,'a').get_attribute('href')
print("The link to the article is", article_link)
```

```python
#Getting the name of the article 
description = article.find_element(By.CLASS_NAME,'name')
print ('the description of the pc: ' + description.text)
```

```python
#Getting the price of the product
price = article.find_element(By.CLASS_NAME,'prc')
print('The price of the pc: ' + price.text)
```

```python
#Getting the old price of the product
old_price = article.find_element(By.CLASS_NAME,'old')
print('The old price of the pc: ' + old_price.text)
```

```python
#Getting the discount
discount = article.find_element(By.XPATH,'.//div[@class="bdg _dsct _sm"]')
print ('the discount is: ' + discount.text)
```

```python
#Getting the rating. Here, if the rating does not exist we give to the rating variable None
try:
    rating = article.find_element(By.XPATH,'.//div[@class="stars _s"]')
except NoSuchElementException:
    rating = None
print('The rating given to this pc is: ',rating)
```

```python
#Getting the image link 
src = article.find_element(By.XPATH,'.//img[@class="img"]').get_attribute("src")
print('The link of the image is: ' + src)
```

#  **Cleaning data**

## Tâche 1 :diviser la description en deux colonnes type et description

```python
# Rappel du DataFrame
articles.head()
```

```python
#Diviser la colonne en deux
#L'argument n dans split() est utilisé pour spécifier le nombre maximum de séparateurs à utiliser lors de la division d'une chaîne de caractères.
description = articles['Decription                                '].str.split(n=1,expand = True)
```

```python
#Attribuer les colonnes crées aux colonnes Type et Name 
articles['Type'] = description[0]
articles['Name'] = description[1]
# Imprimer l'en-tête et confirmer la création des nouvelles colonnes
articles.head()
```

```python
#Supprimer la colonne Decription
#L'argument inplace dans la fonction drop() est un paramètre booléen qui vous permet de modifier le DataFrame en place,
#sans créer une nouvelle copie du DataFrame.
articles.drop('Decription                                ',axis = 1,inplace = True)
#Imprimer l'en-tête du DataFrame
articles.head()
```

```python
# Changer l'ordre des colonnes pour plus de visibilité
articles = articles.reindex(columns=['Type','Name','Price','Old_price', 'Discount','Rating','Link','Image link'])
articles
```

## Tâche 2: convertir les types de Price et Old price en flottants

```python
#Imprimer l'en-tête de notre DataFrame
articles.head()
```

```python
# Enlever le 'Dhs' et le ',' des prix
articles['Price'] = articles['Price'].str.strip('Dhs').str.replace(',','')
articles['Old_price'] = articles['Old_price'].str.strip('Dhs').str.replace(',','')
#Imprimer l'en-tête des deux colonnes avec la méthode iloc qui permet d'accéder aux données
articles.iloc[0:4,[2,3]]
```

```python
#Les types des colonnes
articles.dtypes
```

```python
#Convertir les types des prix en flottant
articles['Price'] = articles['Price'].astype('float')
articles['Old_price'] = articles['Old_price'].astype('float')
#Afficher les typer des colonnes pour voir si c'est bien transformer
articles.dtypes
```

```python
# Visualiser la répartition des prix et anciens prix
sns.distplot(articles['Price'],hist=True,bins= 20,color='r')
sns.distplot(articles['Old_price'],hist=True,bins= 20,color='g')
plt.xlabel('prices')
plt.title('Distribution of prices')
plt.show()
```

```python
#Imprimer les valeurs max des prix
print('le prix moyen est:        ', articles['Price'].mean())
print('Le prix le plus élevé est:   ', articles['Price'].max())
```

## Tâche 3 : convertir le 'Discount' en proportion (flotteur)

```python
#Imprimer l'en-tête de DataFrame pour voir ce qu'on va faire
articles.head()
```

```python
#Enlever le '%' pour la colonne de Discount
articles['Discount'] = articles['Discount'].str.strip('%')
#Verfier si le changement a été bien fait
articles.head()
```

```python
 #Changer le type de Discount en float
articles['Discount'] = articles['Discount'].astype('float')
#Imprimer le type de la colonne 'Discount'
articles['Discount'].dtype
```

```python
#Changer les valeurs de Discount en proportions
articles['Discount'] = articles['Discount']/100
#Verfier si le changement a été bien fait
articles['Discount'].head()
```

## tâche 4 : Traiter les données manquantes dans les colonnes `Discount` et `Old price`

```python
#Lorsque le Discount est NAN cela veut dire qu'il y a pas de remise, donc on remplace les NAN par 0
articles = articles.fillna({'Discount' : 0})
#Afficher les Pcs dont les remises valent 0 dans l'en-tête du DataFrame
articles.sort_values(by = 'Discount').head()
```

```python
# Lorsque il y a pas de remise, l'ancien prix vaut le prix normale
articles = articles.fillna({'Old_price' : articles['Price']})
#Afficher les modifications 
articles.sort_values(by = 'Discount').head()
```

```python
#Vérifier la disparition des NAN pour le Discount et le Old price
articles.info()
```

```python
# Visualiser la répartition des prix et anciens prix
sns.distplot(articles['Price'],hist=True,bins= 20,color='r')
sns.distplot(articles['Old_price'],hist=True,bins= 20,color='g')
plt.xlabel('prices')
plt.title('Distribution of prices')
plt.show()
```

## tâche 5 : Traiter les données manquantes dans la colonne Rating

Comme on ne peut pas supprimer les lignes pour lesquelles le rating n'existe pas (Beaucoup de valeurs manquantes: 42) et on peut pas les prédire, on crée une autre colonne `is Rated` qui contient 1 si le PC est évalué et 0 sinon

```python
# Créer la colonne is Rated
isRated = np.where(articles['Rating'].isna() == True, 0, 1)
articles['is Rated'] = isRated
#Afficher l'en-tête de Data
articles.head()
```

 Tâche 6 : Catégoriser les PCs selon leurs prix

```python
#Ajout d'une nouvelle colonne dans laquelle on spécifie si le PC est abordable ou cher
categorie = np.where(articles['Price']>700, 'Cher', 'Abordable')
articles['Categories'] = categorie
articles.head()
```
